```python
import pandas as pd

# 1. Let's create dummy Mutual Fund data that skips Saturday & Sunday
data = {
    'Date': ['2026-05-28', '2026-05-29', '2026-06-01'],  # Notice May 30 & 31 are completely missing!
    'NAV': [100.0, 101.5, 102.0]
}
df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

print("--- BEFORE FIXING (Missing Weekends) ---")
print(df)

# 2. Fix: Create a continuous daily calendar range (including weekends)
full_calendar = pd.date_range(start=df.index.min(), end=df.index.max(), freq='D')

# 3. Fix: Reindex the data to the full calendar, and use Forward Fill (ffill)
df_cleaned = df.reindex(full_calendar)
df_cleaned['NAV'] = df_cleaned['NAV'].ffill()
df_cleaned.index.name = 'Date'

print("\n--- AFTER FIXING (Weekends filled with Friday's price) ---")
print(df_cleaned)
```
